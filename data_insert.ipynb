```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.engine import URL

# ---------------------------------------------------------
# 1. 데이터 불러오기
# ---------------------------------------------------------
file_name = "processed_vehicle_registrations.csv"
df = pd.read_csv(file_name)

print(f"✅ 총 {len(df)}건의 데이터를 준비했습니다.\n")

# ---------------------------------------------------------
# 2. 데이터베이스 접속 정보 설정 (본인 환경에 맞게 수정 필수!)
# ---------------------------------------------------------
DB_USER = 'root'            # MySQL 아이디 (기본값: root)
DB_PASS = '비밀번호'            # MySQL 비밀번호 (비밀번호가 없다면 '' 로 비워두세요)
DB_HOST = '127.0.0.1'       # 데이터베이스 서버 주소 (내 컴퓨터인 경우 127.0.0.1)
DB_PORT = 3306              # MySQL 기본 포트 번호
DB_NAME = 'vehicle_db_year'      # 생성해둔 데이터베이스 이름

# 안전한 방식으로 데이터베이스 연결 URL 생성 (특수문자 및 빈 비밀번호 오류 방지)
db_url = URL.create(
    drivername="mysql+pymysql",
    username=DB_USER,
    password=DB_PASS if DB_PASS else None,
    host=DB_HOST,
    port=DB_PORT,
    database=DB_NAME
)

print(f"👉 접속 시도 중... (Host: {DB_HOST}, Port: {DB_PORT})")

# ---------------------------------------------------------
# 3. 데이터베이스 연결 및 데이터 밀어넣기
# ---------------------------------------------------------
try:
    # 엔진 생성
    engine = create_engine(db_url)
    
    # DB 연결 테스트
    with engine.connect() as conn:
        print("✅ 데이터베이스 서버 연결 성공!\n")
        
    # 데이터 삽입 (if_exists='append' 로 기존 테이블에 데이터 추가)
    print("⏳ 데이터를 저장하는 중입니다...")
    df.to_sql(name='vehicle_registrations', 
              con=engine, 
              if_exists='append', 
              index=False)
              
    print("🎉 모든 데이터가 성공적으로 저장되었습니다!")

except Exception as e:
    print(f"\n❌ 오류 발생:\n{e}")
    print("\n💡 해결 팁:")
    print("1. MySQL 서버가 켜져 있는지 확인해주세요.")
    print("2. DB_USER와 DB_PASS가 정확한지 확인해주세요.")
    print("3. 'vehicle_db' 데이터베이스가 미리 생성되어 있어야 합니다.")
```
